```python
# Bibliot√©cas utilizadas
import pandas as pd
```

```python
# Import de dados e Path
loc_1 = r'/content/customer_data.csv'
loc_2 = r'/content/payment_data.csv'

# Criando o primeiro DF baseado nos clientes.
df_customer = pd.read_csv(loc_1)

df_customer.head()
```

```python
df_customer.info()
```

```python
print(f'Existem {df_customer.isna().sum().sum()} valores NaN no dataframe.')
print()
print(f'Existe {df_customer.duplicated().sum()} duplicatadas no dataframe')
```

```python
# Criando o segundo DF baseado nos pagamentos.
df_payment = pd.read_csv(loc_2)
df_payment.head()
```

```python
df_payment.info()
```

```python
print(f'Existem {df_payment.isna().sum().sum()} valores NaN no dataframe.')
print()
print(f'Existe {df_payment.duplicated().sum()} duplicatas no dataframe')
```

```python
# Remover duplicatas do dataframe df_payment com base no ID do cliente
df_payment_unique = df_payment.drop_duplicates(keep='first')
# Concatenar os dataframes df_customer e df_payment_unique com base no ID do cliente
df_combined = pd.merge(df_customer, df_payment_unique, on='id', how='inner')

# Habilitando a visualiza√ß√£o de todas as colunas.
pd.set_option('display.max_columns', 35 )

# Verificar o novo dataframe combinado
df_combined.head()
```
