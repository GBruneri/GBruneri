```python
# importando a biblioteca pandas para realizar a manipulação do dataframe.
import pandas as pd

# Criando uma variável de caminho.
caminho = "/content/drive/MyDrive/Code/Myrep/Car_sales/Datasets/CarPrice_Assignment.csv"
# Carregando e exibindo o dataframe
db= pd.read_csv(caminho)
db.head()
```

#### Como temos uma grande variedade de variáveis, é necessário realizar algumas modificações uma vez que precisamos que elas sejam do tipo numérica.

```python
# Estou removendo essas três colunas (Variáveis), pois são desnecessária para nossa finalidade.
db.drop(['car_ID', 'symboling', 'CarName'], axis = 1, inplace = True)
```

```python
# Investigando resultados das variáveis que apresentam tipo "object".
print(db['fueltype'].unique())
print(db['aspiration'].unique())
print(db['carbody'].unique())
print(db['drivewheel'].unique())
print(db['enginelocation'].unique())
print(db['enginetype'].unique())
print(db['cylindernumber'].unique())
print(db['doornumber'].unique())
print(db['fuelsystem'].unique())
```

```python
#Selecionando todas as variáveis catégorias
db_cat = db.select_dtypes(include=['object'])
```

```python
!pip install category_encoders
# Importando um encoder para as variáveis categóricas.
import category_encoders as ce

# Aplicando a função para classificar cada categoria de forma numérica.
encoder = ce.OrdinalEncoder(cols=db_cat)
db = encoder.fit_transform(db)
db
```

```python
# Verificando se todos são do tipo numérico.
db.info()
```

```python
# Criando um caminho para salvar o dataframe.
salva = "/content/drive/MyDrive/Code/Myrep/Car_sales/Datasets/dbr.csv"

# Salve o DataFrame
db.to_csv(salva, index=False)
```
