```python
# importando a biblioteca pandas para realizar a manipulaÃ§Ã£o do dataframe.
import pandas as pd

# Criando uma variÃ¡vel de caminho.
caminho = "/content/drive/MyDrive/Code/Myrep/Car_sales/Datasets/CarPrice_Assignment.csv"
# Carregando e exibindo o dataframe
db= pd.read_csv(caminho)
db.head()
```

#### Como temos uma grande variedade de variÃ¡veis, Ã© necessÃ¡rio realizar algumas modificaÃ§Ãµes uma vez que precisamos que elas sejam do tipo numÃ©rica.

```python
# Estou removendo essas trÃªs colunas (VariÃ¡veis), pois sÃ£o desnecessÃ¡ria para nossa finalidade.
db.drop(['car_ID', 'symboling', 'CarName'], axis = 1, inplace = True)
```

```python
# Investigando resultados das variÃ¡veis que apresentam tipo "object".
print(db['fueltype'].unique())
print(db['aspiration'].unique())
print(db['carbody'].unique())
print(db['drivewheel'].unique())
print(db['enginelocation'].unique())
print(db['enginetype'].unique())
print(db['cylindernumber'].unique())
print(db['doornumber'].unique())
print(db['fuelsystem'].unique())
```

```python
#Selecionando todas as variÃ¡veis catÃ©gorias
db_cat = db.select_dtypes(include=['object'])
```

```python
!pip install category_encoders
# Importando um encoder para as variÃ¡veis categÃ³ricas.
import category_encoders as ce

# Aplicando a funÃ§Ã£o para classificar cada categoria de forma numÃ©rica.
encoder = ce.OrdinalEncoder(cols=db_cat)
db = encoder.fit_transform(db)
db
```
