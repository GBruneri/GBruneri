```python
# Importando bibliotecas necessárias.
import pandas as pd

# Instânciando o caminho dos dados.
dados = '/content/drive/MyDrive/Code/Myrep/Churn_Clientes/Datasets/Customer-Churn.csv'

#Criando o Dataframe
df = pd.read_csv(dados)

# Habilitando a visualização de todas as colunas.
pd.set_option('display.max_columns', 35 )
df.head()
```

```python
df.info()
```

```python
# Criando um dicionário para transformar variáveis.
sim_e_nao = { 'Sim': 1,
              'Nao': 0,
              'SemServicoDeInternet' : 0}
df_1 = df[['Conjuge','Dependentes','TelefoneFixo','SegurancaOnline','BackupOnline',
         'SeguroNoDispositivo','SuporteTecnico','TVaCabo','StreamingDeFilmes','PagamentoOnline','Churn']].replace(sim_e_nao)

df_1.info()
```

```python
# Utilizando get dummies para categorizar as demais variáveis.
df_2 = pd.get_dummies(df.drop(['Conjuge','Dependentes','TelefoneFixo','SegurancaOnline','BackupOnline',
         'SeguroNoDispositivo','SuporteTecnico','TVaCabo','StreamingDeFilmes','PagamentoOnline','ContaMensal','Churn'], axis= 1))

# Encontre as colunas booleanas
boolean_columns = df_2.select_dtypes(include=['bool']).columns

# Use df.loc para aplicar a conversão apenas nas colunas booleanas
df_2.loc[:, boolean_columns] = df_2[boolean_columns].astype(int)

# Concatenando os dados categorizados.
df_3 =pd.concat([df_1, df_2], axis= 1)
df_3.info()
```

```python
# Criando um caminho para salvar o dataframe.
salva ='/content/drive/MyDrive/Code/Myrep/Churn_Clientes/Datasets/db_tratado.csv'

# Salve o DataFrame
df_3.to_csv(salva, index=False)
```
